Due to the nature of the dataset, Support Vector Machine (SVM) would be a good model choice. It is known to perform well and avoid overfitting when there are a lot of features, a large set of data, and the data is unbalanced. I trained an SVM model with a linear kernel, taking over 40 minutes and was unable to correctly predict any transactions as fraud (MCC = 0.0), similar to logistic regression. I chose to change the kernel to RBF to enhance the decision boundary. This model did not finish training 40 minutes in. Due to hardware/time constraints I abandoned this approach and chose to train a K-Nearest Neighbors. I chose to use a clustering approach for training to see how it will impact classification on our dataset. I hypothesized that the performance would be increase due to the limited fraudulent transaction seen being grouped together. This algorithm yielded performance than logistic regression and Linear SVM. The precision and recall were 0.840 and 0.798 respectively. More meaningfully, the MCC was about 0.637.

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
model_rf = Fraud_Detector_Model(rf)
rf_trained = model_rf.train(filename)
```

```python
rf_trained
```

```python
model_rf.test()
```

The last model I chose to train is Random Forest model. These models also are known to perform well on unbalanced data with many features. The random forest resulted in much better (17.9% increase in MCC from kNN) performance than logistic regression, Linear SVM. The precision and recall,  and accuracy were 0.972 and 0.853 respectively. More meaningfully, the MCC was about 0.816. According to the confusion matrix, 694 true positives were detected and 290 missed. This is contrasted with all 984 positives being missed in the logistic regression model. The random forest model yields the best performance of the 3 models and is reasonably able to be trained in our system environment. The reasonably fast training time will allow for model data updates to be made to the system and yield accurate predictions.
